### Change the datatypes of columns

```python
apps_df['Released'] = pd.to_datetime(apps_df['Released'])
apps_df['Last Updated'] = pd.to_datetime(apps_df['Last Updated'])
```

The column "Size" of the apps_df tells the size of the app in KB, MB or GB. So we will convert that to only numbers (in MB).

```python
# convert datatype of column from float to str
apps_df['Size'] = apps_df['Size'].astype(str)

# replace commas with empty string
apps_df['Size'] = apps_df['Size'].str.replace(',', '')

apps_df['Size']
```

```python
# define a function to convert a single value to a number
def convert_value(value):
    # check if the value ends with "M"
    if value.endswith('M'):
        # remove the "M" character and returns value of MB in float
        return float(value[:-1])
    # check if the value ends with "k"
    elif value.endswith('k'):
        # remove the "k" character and converts & returns the value of KB in MB in float
        return float(value[:-1]) / 1000
    # if the value is 'Varies with device' returns 0
    elif value.endswith('device'):
        return '0'
    elif value.endswith('G'):
        # remove the "G" character and converts & return the value of GB in MB in float
        return float(value[:-1]) * 1000
    else:
        return float(value)

# apply the function to the "Size" column
apps_df['Size'] = apps_df['Size'].apply(convert_value)

# print the "Size" column
print(apps_df['Size'])
```

Now all the values are in Megabytes.

```python
# Again convert the "Size" column to float64
apps_df['Size'] = pd.to_numeric(apps_df['Size'],errors='coerce')
```

```python
size = apps_df['Size']

# define a function to create 5 categories of size type
def size_cat(size):
    if size == 0.0:
        return 'Size not provided'
    elif size > 0.0 and size <= 100.0:
        return 'Under 100 MB'
    elif size > 100.0 and size <= 500.0:
        return 'Between 100 and 500 MB'
    elif size > 500.0 and size <= 1024.0:
        return 'Between 500 MB and 1 GB '
    else:
        return 'More than 1 GB'
```

```python
apps_df['Size Type']= apps_df["Size"].apply(size_cat)
```

```python
apps_df['Size Type'].unique()
```

Installs column is of the object type and has '+' symbol we have to remove it and convert it to init.

```python
# Remove the Symbols
apps_df["Installs"] = apps_df["Installs"].str.split("+").str[0]

# Replace the commas"," into install columns
apps_df["Installs"].replace(",", "", regex=True, inplace=True)

```

```python
# Filling Nan with 0
apps_df['Installs'] = apps_df['Installs'].fillna('0').astype(int)
apps_df['Installs'].unique()
```

```python
# converting it to init type
apps_df["Installs"] = apps_df["Installs"].astype(int)
apps_df['Installs']
```

**Making categories for pricing**
- free 0
- low >0 and <= 10
- medium >10 and <=35
- high >35 and <= 100

```python
pr = apps_df['Price']
def make_cat(pr):
    if pr == 0.0:
        return 'Free'
    elif pr > 0.0 and pr <= 10.0:
        return 'Low'
    elif pr > 10.0 and pr <= 35.0:
        return 'Medium'
    elif pr > 35.0 and pr <= 100.0:
        return 'High'
    else:
        return 'Expensive'
```

## Asking and Answering Questions

**1. Which are the top 10 installed categories from google play store?**

```python
categories
```

```python
top_installs = apps_df.groupby(['Category'])[['Installs']].sum()
top_installs
```

```python
category_top_installs = pd.merge(categories, top_installs, on='Category').sort_values(by='Installs', ascending=False).reset_index()
category_top_installs
```

**6. What is the distribution of pricing among the paid apps?**

```python
pr = apps_df[apps_df['Pricing Type']!='Free']['Pricing Type'].value_counts()
pr
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

perc = [f'{i/pr.values.sum()*100:5.2f}%' for i in pr.values]
lbl = [f'{el[0]} = {el[1]}' for el in zip(pr.index, perc)]

# Create the treemap
squarify.plot(sizes=pr.values, label=lbl, alpha=.8, color=['#f94144','#f8961e','#f9c74f','#90be6d'], ax=ax)

# Remove axes ticks and labels
plt.axis('off')

# Add a title
ax.set_title('Distribution of Pricing Categories')

# Show the plot
plt.show()
```

- Here from the treemap we can see that the apps with low price range are about 93.53% and the expensive apps are only 0.41%.
-There is a large difference between low price and other three price range.
